```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import integrate
```

# Part 1

```python
#loading data
time, vel = np.loadtxt("droptower_vdata.txt", unpack = True)

dist = scipy.integrate.cumtrapz(vel, time, initial = 0)
acc = np.diff(vel)
```

```python
#plotting position-time
plt.plot(time, dist)
plt.xlabel("time (s)")
plt.ylabel("position (m)")
plt.title("Position vs. Time")
plt.show()
```

```python
#plotting vel-time
plt.plot(time, vel)
plt.xlabel("time (s)")
plt.ylabel("velocity (m/s)")
plt.title("Velocity vs. Time")
plt.show()
```

```python
#plotting acc-time
plt.plot(time[0:10], acc)
plt.xlabel("time (s)")
plt.ylabel("acceleration (m/s^2)")
plt.title("Acceleration vs. Time")
plt.show()
```

# Part 2

```python
#getting rid of negative acceleration
pos_acc = []
for e in acc:
    if e > 0 or e == 0:
        pos_acc.append(e)

#printing avarage acceleration
print("The average acceleration of the cab when it is accelerating upwards is",round(np.mean(pos_acc),3),"m/s^2.")
```
